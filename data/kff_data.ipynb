```r
# Preliminaries --------------------------------------------------------
if (!require("pacman")) install.packages("pacman")
pacman::p_load(tidyverse, ggplot2, dplyr, lubridate, haven)
```

```r
# Preliminaries --------------------------------------------------------
kff.dat <- read_csv('input/raw_data__.csv', skip = 2)
```

```r
# Clean KFF data --------------------------------------------------------
kff.final <- kff.dat %>%
  mutate(
    expanded = (`Status of Medicaid Expansion Decision` == "Adopted"),
    date_adopted = mdy(`Expansion Implementation Date`)
  ) %>%
  select(State = Location, expanded, date_adopted)
```

```r
write_rds(kff.final, 'output/medicaid-kff.rds')
```
